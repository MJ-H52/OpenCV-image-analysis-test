# 導入需要的模組. Import required modules.

```python
import cv2
import numpy as np
from numpy import array, uint8

```

# 載入圖片. Load images.


```python
# img = cv2.imread("twcoin.jpg")
img = cv2.imread("S__5775901_0.jpg")
cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()


cv2.destroyAllWindows()
```

# 設定縮放比例. Set scaling

```python
scale_percent = 20  # 原始圖片的 60%
width = int(img.shape[1] * scale_percent / 100)
height = int(img.shape[0] * scale_percent / 100)
dim = (width, height)

img = cv2.resize(img, dim, interpolation = cv2.INTER_AREA)# 縮放圖片


cv2.imshow("resize_img", img)
cv2.waitKey(0)
```

# copy 變更完成的照片. copy the changed photo.

```python
image_copy = img.copy()

cv2.imshow("image_copy", image_copy)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 變更為Gray. Change to Gray.

```python

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

cv2.imshow("cvtColor_img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 照片經過模糊化. Photos are blurred.

```python
img = cv2.GaussianBlur(img, (5, 5) , 1)

cv2.imshow('GaussianBlur_img', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 照片經過二值化. Photos are binarized.

```python
ret , thresh = cv2.threshold(img, 120, 255, cv2.THRESH_BINARY)

cv2.imshow('threshold_img', thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 找尋輪廓. Find outlines .

```python
contours , _ = cv2.findContours(thresh , cv2.RETR_TREE , cv2.CHAIN_APPROX_NONE)

print(contours)
```

# 取得每一個區域的面積. Get the area of ​​each region.

```python

area = {}
for i in range(len(contours)):
    cnt = contours[i]
    ar = cv2.contourArea(cnt)
    area[i] = ar
print(area)
```

# 進行面積排序. Sort by area.

```python
srt = sorted(area.items(), key= lambda x : x[1], reverse= True)
results = np.array(srt).astype("int")
print(results)
```

# 選取大於設定的面積. Select an area larger than the set area.

```python
num = np.argwhere(results[: , 1] > 4000).shape[0]

print(num)
```

# 畫出框選的區域. Draw the selected area.

```python
for i in range(1 , num):
    image_copy = cv2.drawContours(image_copy , contours , results[i , 0] ,(0 , 255 , 0) , 3)
    # image_finially = cv2.drawContours(image_copy , contours , results[i , 0] , (0 , 255 , 0) , 3)
print("Number of coins is " , num - 1)

cv2.imshow('finially', image_copy)
# cv2.imshow('finially', image_finially)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
